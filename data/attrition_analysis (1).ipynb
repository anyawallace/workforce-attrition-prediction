# Workforce Attrition Analysis

## Problem Statement
Employee attrition creates staffing gaps, increases training costs,
and reduces organizational readiness.

The goal of this analysis is to explore workforce characteristics
and build a foundation for predicting employee attrition using
publicly available data.

## Dataset
This project uses the IBM HR Analytics Employee Attrition dataset,
a commonly used public dataset for workforce analytics.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display settings
pd.set_option("display.max_columns", None)
```

```python
# Load dataset directly from a public URL
url = "https://raw.githubusercontent.com/IBM/employee-attrition-aif360/master/data/emp_attrition.csv"
df = pd.read_csv(url)

# Show basic info
df.shape
```

```python
# Preview the data
df.head()
```

```python
# Check attrition counts
df["Attrition"].value_counts()
```

```python
# Plot attrition distribution
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

## Initial Observations
- The dataset contains 1,470 employees with a binary attrition outcome.
- Attrition is relatively rare, indicating a class imbalance.
- This mirrors real-world workforce attrition problems and
  informs later modeling choices.
